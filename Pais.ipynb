1. Qual o nome do país que possui mais cidades registradas?

```python
country_dict = {
    1: "India",
    14: "Australia",
    30: "Brazil",
    37: "Canada",
    94: "Indonesia",
    148: "New Zeland",
    162: "Philippines",
    166: "Qatar",
    184: "Singapure",
    189: "South Africa",
    191: "Sri Lanka",
    208: "Turkey",
    214: "United Arab Emirates",
    215: "England",
    216: "United States of America",
}


df['country'] = df['Country Code'].map(country_dict)
df_country = df.drop(columns=['Country Code'])
df_country.sample(2)
```

```python
df_country.to_csv('df_country.csv', index=False)
```

```python
COLORS = {
    "3F7E00": "darkgreen",
    "5BA829": "green",
    "9ACD32": "lightgreen",
    "CDD614": "orange",
    "FFBA00": "red",
    "CBCBC8": "darkred",
    "FF7800": "darkred",
}

df['color_name'] = df['Rating color'].map(COLORS)
df_country = df.drop(columns=['Rating color'])
df_country.sample(1000)
```
