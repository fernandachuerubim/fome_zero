```python
#Libraries
from haversine import haversine
import plotly.express as px
import plotly.graph_objects as go

#bibliotecas necessárias
import folium
import pandas as pd
import streamlit as st
from PIL import Image
```

```python
#================
#import dataset
#================
df = pd.read_csv(r'zomato.csv')
```

```python
1. Qual o nome da cidade que possui mais restaurantes registrados?
```

```python
nome = df.loc[:, ['City', 'Restaurant Name']].groupby('City').count().sort_values('Restaurant Name', ascending=False)
nome
```

```python
2. Qual o nome da cidade que possui mais restaurantes com nota média acima de 4?
```

```python
filtro = df['Aggregate rating'] > 4
df2 = df.loc[filtro, :]

df2.loc[:, ['City', 'Restaurant Name']].groupby('City').count().sort_values('Restaurant Name', ascending=False)
```

```python
3. Qual o nome da cidade que possui mais restaurantes com nota média abaixo de 2.5?
```

```python
filtro = df['Aggregate rating'] < 2.5
df2 = df.loc[filtro, :]

df2.loc[:, ['City', 'Restaurant Name']].groupby('City').count().sort_values('Restaurant Name', ascending=False)
```

```python
4. Qual o nome da cidade que possui o maior valor médio de um prato para dois?
```

```python
df.loc[:, ['City','Average Cost for two']].groupby('City').mean().sort_values('Average Cost for two', ascending=False)
```

```python
5. Qual o nome da cidade que possui a maior quantidade de tipos de culinária distintas?
```

```python
df.loc[:, ['City', 'Cuisines']].groupby('City').nunique().sort_values('Cuisines', ascending=False)
```

```python
6. Qual o nome da cidade que possui a maior quantidade de restaurantes que fazem reservas?
```

```python
filtro = df['Has Table booking'] == 1
df2 = df.loc[filtro, :]

df2.loc[:, ['City', 'Restaurant Name']].groupby('City').count().sort_values('Restaurant Name', ascending=False)
```

```python
7. Qual o nome da cidade que possui a maior quantidade de restaurantes que fazem entregas?
```

```python
filtro = df['Has Online delivery'] | df['Is delivering now']
df2 = df.loc[filtro, :]

df2.loc[:, ['City', 'Restaurant Name']].groupby('City').count().sort_values('Restaurant Name', ascending=False)
```
